# Part 2

```python
deltas = []
with open('input.txt') as infile:
    for l in infile:
        l = l.strip()
        v = int(l[1:])
        if l[0] == '-':
            v = 0 - v
        deltas.append(v)

def loop(vals):
    while True:
        for v in vals:
            yield v
        
seen = {0}
total = 0
for v in loop(deltas):
    total += v
    if total in seen:
        break
    seen.add(total)
total
```
