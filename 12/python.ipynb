```python
from collections import deque

def trim(dq, idx):
    # assumes dq is never empty
    while dq[0] == '.':
        idx += 1
        dq.popleft()
    while dq[-1] == '.':
        dq.pop()
    for _ in range(3):
        dq.appendleft('.')
        dq.append('.')
    idx -= 3
    return idx

dq = deque(state0)
idx = 0
for x in range(2000):
    idx = trim(dq, idx)
    state = ''.join(dq)
    print(x, len(state), idx, score(dq, idx))
    dq = deque(
        rules.get(state[i:i + 5], '.')
        for i in range(0, len(state) - 4)
    )
    idx += 2
(50 * 10^9 - 118) * 69 + 10410
```

```python
def score(dq, idx):
    total = 0
    for i, pot in enumerate(dq):
        if pot == '#':
            total += i + idx
    return total
```
